```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data used for the descriptive analysis 
greggs_2015 = yf.download("GRG.L", start = "2015-01-02", end = "2018-12-29")

dominos_2015 = yf.download("DOM.L", start = "2015-01-02", end = "2018-12-29")

tesco_2015 = yf.download("TSCO.L", start = "2015-01-02", end = "2018-12-29")

sainsburys_2015 = yf.download("SBRY.L", start = "2015-01-02", end = "2018-12-29")
```

# Dominos

```python
dominos_2015.agg(["min", "max", "mean", "std"])
```

```python
#Open and close

fig, ax = plt.subplots(figsize=(30,15))
ax.plot(dominos_2015['Open'])
ax.plot(dominos_2015['Close'])
ax.plot(dominos_2015['Adj Close'])

plt.title("Dominos", size=30)
plt.xlabel("Dates", fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(["Open", "Close", "Adj Close"],  prop={"size":30})

plt.show()
```

```python
dominos_2015[["Open", "Close", "Adj Close"]].agg(['min', 'max', 'mean', 'std'])
```

```python
#low and High

fig, ax = plt.subplots(figsize=(30,15))
ax.plot(dominos_2015['Low'])
ax.plot(dominos_2015['High'])

plt.title("Dominos", size=30)
plt.xlabel("Dates", fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.legend(["Low", "High"], prop={'size':30})

plt.show()
```

```python
dominos_2015[["Low", "High"]].agg(['min', 'max', 'mean', 'std'])
```

```python
#Volume

fig, ax = plt.subplots(figsize=(30,15))
ax.plot(dominos_2015['Volume'])

plt.title("Dominos", size=30)
plt.xlabel("dates", fontsize=30)
plt.ylabel("Volume", fontsize=30)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)


plt.show()
```

```python
dominos_2015[["Volume"]].agg(['min', 'max', 'mean', 'std'])
```
